```python
import pandas as pd
df = pd.read_csv('homeprices.csv')
df
```

```python
# creating dummy variables columns
dummies = pd.get_dummies(df.town,drop_first=True)
dummies
```

```python
# merging dummy columns with df
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
# dropping town column and any one dummy variable column (to avoid dummy variable trap)

final = merged.drop(['town','west windsor'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop('price', axis='columns') # x is area, monroe township, robinsville
x
```

```python
y = final.price
y
```

```python
model.fit(x,y)
```

```python
# lets see price of 2800 sq.ft. home in robinsville i.e. (0,1)
model.predict([[2800, 0, 1]])
```

```python
# lets see price of 3400 sq.ft. home in west windsor i.e. (0,0)
model.predict([[3400, 0, 0]])
```

# Checking Accuracy of the model

```python
model.score(x,y) # 95.7% accurate
```
